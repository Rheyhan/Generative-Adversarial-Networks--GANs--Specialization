```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2

def df(x):
    return 2 * x

def steepest_descent(initial_x, learning_rate, num_iterations=1000, tol=1e-6):
    x = initial_x
    temp=[] 
    
    for i in range(num_iterations):
        gradient = df(x)
        x = x - learning_rate * gradient
        print(f"Iteration {i + 1}: x = {x}, f(x) = {f(x)}")
        temp.append([i+1, x, f(x)])
        if gradient<tol: break
        
    return x, temp

# Initial values
initial_x = 5
learning_rate = 0.1
num_iterations = 1000

# Run steepest descent
final_x, hist = steepest_descent(initial_x, learning_rate, num_iterations)

print("\nSteepest Descent Result:")
print(f"Initial x: {initial_x}")
print(f"Final x: {final_x}")
print(f"Final f(x): {f(final_x)}")
print(f'Final epoch: {hist[-1][0]}')
```

```python
z=np.array(hist)
plt.plot(z[:,0], z[:,2])
plt.grid("on", alpha=0.3)
plt.ylabel("Iteration")
plt.xlabel("F(x)")
```

```python
import numpy as np
from scipy.optimize import minimize_scalar

def f(x):
    return x**2

def grad_f(x):
    return 2 * x

def conjugate_gradient_fr(f, x0, tol=1e-6, max_iter=1000):
    x = x0
    fx = f(x)
    gradx = grad_f(x)
    d = -gradx
    iteration = 0
    
    while iteration < max_iter and gradx > tol:
        optimize_result = minimize_scalar(lambda a: f(x + a * d), bounds=(0, 1), method='bounded')
        alpha = optimize_result.x
        
        x = x + alpha * d
        gradx_new = grad_f(x)
        
        beta = gradx_new**2 / gradx**2
        d = -gradx_new + beta * d
        
        fx = f(x)
        gradx = gradx_new
        
        iteration += 1
    
    return {'x': x, 'fx': fx, 'iter': iteration}

# Example usage:
result = conjugate_gradient_fr(f, 5)

# Print the result
print("Conjugate Gradient Result:")
print(f"Initial x: 5")
print(f"Final x: {result['x']}")
print(f"Final f(x): {result['fx']}")
print(f"Iterations: {result['iter']}")
```
